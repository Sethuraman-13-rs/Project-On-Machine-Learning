```python
# import numerical python 
import numpy as np

# import the pandas 
import pandas as pd

# import the seaborn
import seaborn as sns

# import the matplotlib
import matplotlib.pyplot as plt
```

```python
# load the data
df_house_price=pd.read_csv('House_Price1.csv')
```

```python
# show the first five observation of the data 
df_house_price.head()
```

```python
Dictionary={
    'price':'Price',
    'resid_area':'Resident_Area',
    'air_equal':'Air_Equal',
    'room_num':'Room_Num',
    'age':'Age',
    'distance':'Distance',
    'teachers':'Teachers',
    'poor_pop':'Poor_Pop',
    'airport':'Airport',
    'n_hos_beds':'N_Hos_Beds',
    'n_hot_rooms':'N_Hot_Rooms',
    'waterbody':'Waterbody',
    'rainfall':'Rainfall',
    'bus_ter':'Bus_Ter',
    'sold':'Sold'
}
```

```python
# perform the rename action
df_house_price.rename(columns=Dictionary)
```

```python
# Lets Treat the Null Value in Waterbody Column 

df_house_price.waterbody.fillna('River', inplace = True)
```

```python
# N_Hospital_Beds,we have less than ten percent data missing
df_house_price.dropna(inplace=True)
```

```python
df_target = df_house_price['Sold']
```

```python
df_feature=df_house_price.drop('Sold',axis=1)
```

```python
# segregate the data based on datatype
df_num=df_house_price.select_dtypes(include='number')
df_cat=df_house_price.select_dtypes(include='object')
```

```python
# use dummy encoding 
df_encoded=pd.get_dummies(df_cat)
# reset the index 
df_encoded=df_encoded.reset_index()
```

```python
# import the standard scaler
from sklearn.preprocessing import StandardScaler 

# initiate the standard scalar 
ss = StandardScaler()
```

```python
# Scaled the value present in numerical data frame
scaled = ss.fit_transform(df_num)
```

```python
# Numerate the data into the dataframe
scaled_df=pd.DataFrame(scaled,columns=df_num.columns)
```

```python
# Reset the Index 
scaled_df=scaled_df.reset_index()
```

```python
# concat encoded the variable and scaled variable
df_independent = pd.concat([df_encoded,scaled_df],axis=1) 
```

```python
# import the train test split
from sklearn.model_selection import train_test_split
```

```python
# split the data 
x_train,x_test,y_train,y_test=train_test_split(df_independent,df_target,test_size=0.2)
```

### Decision Tree Classified

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report

#create a decision tree classifier
tree_model=DecisionTreeClassifier()

# Fit the model to the training data
tree_model.fit(x_train,y_train)

# predict on the testing data
predictions = tree_model.predict(x_test)

# calculate classification report
class_report = classification_report(y_test,predictions)
print(class_report)
```
